```python
import pandas as pd
import json
```

```python
df = pd.read_json("data.json")
```

```python
df
```

```python
#social_D all variables
social_D_90_days = []
social_D_cum_abs = []
cumul_z = []


#price_D
price_D_volume = []
price_D_last = []

date = []
```

```python
for i in range(0,len(df)):
    social_D_90_days.append(df['social_D'][i]['90day_bull_prop_rollingz_60'])
    social_D_cum_abs.append(df['social_D'][i]['cumulative_abs'])
    
    #price_D
    price_D_volume.append(df['price_D'][i]['volume'])
    price_D_last.append(df['price_D'][i]['last'])
    date.append(df['price_D'][i]['date'])
    
    #cumulative
    cumul_z.append(df['social_D'][i]['cumulative_z'])
    
    
```

```python
#social_D
print("Min of 90day bull prop rolling_z in social_D is:",min(social_D_90_days))
print("Max of 90day bull prop rolling_z in social_D is:",max(social_D_90_days))
print("Avg of 90day bull prop rolling_z in social_D is:",sum(social_D_90_days)/len(df))

print("--------------------------------------------------------------------------------")

print("Min of cumulative_abs in social_D is:",min(social_D_cum_abs))
print("Max of cumulative_abs in social_D is:",max(social_D_cum_abs))
print("Avg of cumulative_abs in social_D is:",sum(social_D_cum_abs)/len(df))
print("--------------------------------------------------------------------------------")

#price_D
print("Min of volume in price_D is:",min(price_D_volume))
print("Max of volume in price_D is:",max(price_D_volume))
print("Avg of volume in price_D is:",sum(price_D_volume)/len(df))
print("--------------------------------------------------------------------------------")

print("Min of last in price_D is:",min(price_D_last))
print("Max of last in price_D is:",max(price_D_last))
print("Avg of last in price_D is:",sum(price_D_last)/len(df))


```
