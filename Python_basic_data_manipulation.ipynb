```
import pandas as pd
import json
```

```
df = pd.read_json("data.json")
```

```
df
```

```
social_D_90_day= []
social_D_cum_abs = []
cum_z = []

price_D_volume = []
price_D_last = []
date = []



for i in range(0,len(df)):
    social_D_90_day.append(df['social_D'][i]['90day_bull_prop_rollingz_60'])
    social_D_cum_abs.append(df['social_D'][i]['cumulative_abs'])

    price_D_volume.append(df['price_D'][i]['volume'])
    price_D_last.append(df['price_D'][i]['last'])
    date.append(df['price_D'][i]['date'])
    
    cum_z.append(df['social_D'][i]['cumulative_z'])

print("Minimum 90day_bull_prop_rollingz_60 in social_D is " + str(min(social_D_90_day)))
print("Maximum 90day_bull_prop_rollingz_60 in social_D is " + str(max(social_D_90_day)))
print("Avg 90day_bull_prop_rollingz_60 in social_D is " + str(sum(social_D_90_day)/len(df)))
print("===========================================")
print("Minimum cumulative_abs in social_D is " + str(min(social_D_cum_abs)))
print("Maximum cumulative_abs in social_D is " + str(max(social_D_cum_abs)))
print("Avg cumulative_abs in social_D is " + str(sum(social_D_cum_abs)/len(df)))
print("===========================================")
print("Minimum volume in price_D is " + str(min(price_D_volume)))
print("Maximum volume in price_D is " + str(max(price_D_volume)))
print("Avg volume in price_D is " + str(sum(price_D_volume)/len(df)))
print("===========================================")
print("Minimum last in price_D is " + str(min(price_D_last)))
print("Maximum last in price_D is " + str(max(price_D_last)))
print("Avg last in price_D is " + str(sum(price_D_last)/len(df)))   

```

```
import numpy as np

sorted_index = np.argsort(price_D_last)
```

```
sorted_index
```

```
for i in range(0,3):
    print("Date: " + str(date[sorted_index[i]]))
    print("Price: " + str(price_D_last[sorted_index[i]]))
    print("-----------------------------------")
```

```
df_new = pd.DataFrame(zip(date,social_D_90_day,cum_z,price_D_last), columns = ['date', '90day_bull_prop_rollingz_60', 'cumulative_z', 'last'])
```
